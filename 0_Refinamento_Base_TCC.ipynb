```python
# imports necessary libriries
# importa bibliotecas necessárias

import pandas as pd
from google.colab import drive
import warnings

# imports drive
# importa o drive

drive.mount('/content/drive')

# ignores warnings
# ignora avisos
warnings.filterwarnings('ignore')

```

```python
#imports raw data
#importa base bruta

df = pd.read_csv("/content/drive/MyDrive/TCC/1. Base de Dados/dados_brutos/loan_bruta.csv")
```

```python
#shape of original base
#dimensões da base original

df.shape
```

```python
# proportion of purpose variable
# proporção da variável purpose

a = df['purpose'].value_counts(normalize=True).mul(100).round(2).reset_index()
a['proportion'] = a['proportion'].astype(str) + '%'
display(a)
```

```python
# proportion of loan_status variable
# proporção da variável loan_status

a = df['loan_status'].value_counts(normalize=True).mul(100).round(2).reset_index()
a['proportion'] = a['proportion'].astype(str) + '%'
display(a)
```

```python
#equal quantities of good and bad payers to train dataset
#quantidades iguais de bons e maus pagadores para dataframe de treino

df_treino_fully_paid  = df[(df['loan_status'] == 'Fully Paid')].sample(n  = 175000,  random_state = 5685305)
df_treino_charged_off = df[(df['loan_status'] == 'Charged Off')].sample(n = 175000, random_state = 4965089)

#concats and get the train dataset
#concatena e gera dataframe  de treino dos modelos

df_treino = pd.concat([df_treino_fully_paid,df_treino_charged_off], ignore_index=True)
```

```python
#drop train data to create test data
#exclui dados de treino do conjunto de teste

df_teste  = df.drop(df_treino.index)

#equal quantities of good and bad payers to test dataset
#quantidades iguais de bons e maus pagadores para dataframe de teste

df_teste_fully_paid  = df_teste[(df_teste['loan_status'] == 'Fully Paid')].sample(n =  75000, random_state = 49550)
df_teste_charged_off = df_teste[(df_teste['loan_status'] == 'Charged Off')].sample(n = 75000, random_state = 49650)

#concats and get the test dataset
#concatena e gera dataframe de teste dos modelos

df_teste = pd.concat([df_teste_fully_paid,df_teste_charged_off], ignore_index=True)
```

```python
#validation
#validação
df_treino['loan_status'].value_counts()
```

```python
#validation
#validação
df_teste['loan_status'].value_counts()
```

```python
# salva em csv no drive
#saves in csv file and puts in drive
df_treino.to_csv('/content/drive/MyDrive/TCC/1. Base de Dados/dados_refinados/loan_refinada_treino.csv', index=False)
df_teste.to_csv('/content/drive/MyDrive/TCC/1. Base de Dados/dados_refinados/loan_refinada_teste.csv', index=False)
```
